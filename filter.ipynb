```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = input("Enter Image: ") # in path\filename.ext format
```

```python
pic = Image.open(img)
picar = np.asarray(pic)
```

```python
color = input("Enter Filter Color: ") # only digital(0/1) RGB colors
colDict = {'red':'100,','green':'010' , 'blue':'001', 'yellow':'110', 'cyan':'011', 'purple':'101', 'black':'111'}
```

```python
picard = np.copy(picar)
for i in range(3):
    if int(colDict[color][i]) ==  0:
        picard[:,:,i] = 0
plt.imshow(picard)
```
